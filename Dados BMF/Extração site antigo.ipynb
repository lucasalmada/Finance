```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
import numpy as np
pd.set_option('display.max_columns', 0)
```

```python
#2018 e 2019

df1 = pd.read_excel('Estatistico_pregao_2018-2019.xlsx')

df1 = df1[(df1[['Percentagem de negócios Day Trade']] > 1).all(axis = 1)]
df1 = df1[(df1[['Referência do Resumo']] == 'ELET').all(axis = 1)]
```

```python

#2017
df2 = pd.read_excel('Estatistica_pregao_2017.xlsx')

df2 = df2[(df2[['Percentagem de negócios Day Trade']] > 1).all(axis = 1)]
df2 = df2[(df2[['Referência do Resumo']] == 'ELET').all(axis = 1)]
```

```python
#2016
df3 = pd.read_excel('Estatistica_pregao_2016.xlsx')

df3 = df3[(df3[['Percentagem de negócios Day Trade']] > 1).all(axis = 1)]
df3 = df3[(df3[['Referência do Resumo']] == 'ELET').all(axis = 1)]
```

```python
#2014 e 2015
df4 = pd.read_excel('Estatistico_pregao_2014-2015.xlsx')

df4 = df4[(df4[['Percentagem de negócios Day Trade']] > 1).all(axis = 1)]
df4 = df4[(df4[['Referência do Resumo']] == 'ELET').all(axis = 1)]
```

```python
#2012 e 2013
df5 = pd.read_excel('Estatistico_pregao_2012-2013.xlsx')

df5 = df5[(df5[['Percentagem de negócios Day Trade']] > 1).all(axis = 1)]
df5 = df5[(df5[['Referência do Resumo']] == 'ELET').all(axis = 1)]
```

```python
df_concat = pd.concat([df1,df2,df3,df4,df5]).reset_index(drop = True)
```

```python
df = df_concat.drop(df_concat[['Referência do Resumo', 'Base de calculo de margem - Opções','Contrato objeto da opção',
                         'Delta de margem','Equivalência em moeda corrente', 'Indicador de ajuste anterior definitivo',
                         'Indicador de ajuste atual definitivo', 'Margem por contrato - clientes hedger',
                         'Margem por contrato - clientes normais', 'Mercado', 'Mercadoria', 'Número do vencimento',
                         'Percentagem de negócios Day Trade', 'Percentagem de negócios vinculados', 'Percentagem de negócios vinculados',
                         'Preço de exercício', 'Quantidade da última oferta de compra', 'Quantidade da última oferta de venda',
                         'Quantidade de contratos baixados', 'Quantidade de contratos exercidos', 'Quantidade de contratos travados',
                         'Quantidade de negócios Day Trade', 'Quantidade de negócios de exercício', 'Quantidade de negócios vinculados',
                         'Tamanho do contrato','Tipo de série', 'Valor da moeda de referência anterior','Valor da moeda de referência atual',
                         'Vencimento', 'Volume de exercício do dia', 'Volume de exercício em Dólar',
                              'Data do último negócio', 'Data do vencimento']], axis=1)  
```

```python
df.head()
```

#### Oscilação positiva ou negativa  dia depois

```python
df_oscilacao = df.copy()
```

```python
df_oscilacao['label_osc'] = df_oscilacao['Oscilação'].apply(lambda s: 1 if s > 0 else 0).shift(-1)
```

```python
df_oscilacao.fillna(0, inplace=True)
```

```python
X = df_oscilacao.iloc[:,1:-3]
```

```python
y =df_oscilacao.iloc[:,-1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Random Forests

rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)

rf.fit(X_train, y_train) 

prediction = rf.predict(X_test)

accuracy_score(y_test, prediction)
```

```python
#SVM

from sklearn.svm import SVC
clf = SVC(gamma='auto')
clf.fit(X_train, y_train) 
prediction = clf.predict(X_test)
accuracy_score(y_test, prediction)
```

```python
#Regressão Logística

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train) 
prediction = clf.predict(X_test)
accuracy_score(y_test, prediction)
```

```python
#XGboost

from xgboost.sklearn import XGBClassifier 

xgb_model = XGBClassifier() 

xgb_model.fit(X_train, y_train)

prediction = xgb_model.predict(X_test)

accuracy_score(y_test, prediction)
```
